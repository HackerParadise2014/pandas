# Data

We have data from 2008 on the number of complaints recevied by each PCT. We have the PCT name, code and the region it is located.

```python
# Import data from CSV
pct = pd.read_csv('/vagrant/data/PCT-2008-clean.csv', header=0, index_col=1)
pct
```

## Top 10 PCT By Complaints

```python
top_10_complaints = pct[['Name', 'Complaints']].sort('Complaints', ascending=True)[-10:]
top_10_complaints_name = top_10_complaints.set_index('Name')
top_10_complaints_name
```

## Make A Bar Chart

```python
%matplotlib inline 
top_10_complaints_name.plot(kind='barh', color='red', edgecolor='red', grid=False, legend=False, figsize=(10,7), fontsize=12)
plt.xlabel('Number of Complaints', fontsize=12)
plt.ylabel('Primary Care Trust', fontsize=12)
plt.tick_params(axis='y', which='both', left='off', right='off')    
plt.tick_params(axis='x', which='both', top='off') 
```
