# Data

This is an aggregate data set from the UK Data Service where we have worked out the proporting of the median income for the years 2011, 2001, 1991 each age makes on average

```python
# Import data from csv
income = pd.read_csv('/vagrant/data/income-data.csv', header=0, index_col=0)
income
```

# Visualisation

```python
%matplotlib inline 
from matplotlib.ticker import FuncFormatter 

income.plot(kind='line', ylim=(0, 1.3), grid=False, figsize=(14,8))
plt.xticks(np.arange(min(income.index.values), max(income.index.values)+1, 2))
plt.tick_params(axis='x', which='both', top='off', bottom='off')
plt.tick_params(axis='y', which='both', right='off') 
plt.axhline(1, color='gray')
plt.axes().xaxis.grid()
plt.legend(('2011', '2001', '1991'))

def myfunc(x, pos=0): 
    return '%d %%'%(100*x)

plt.axes().yaxis.set_major_formatter(FuncFormatter(myfunc)) 
plt.ylabel('Income as percentage of median', fontsize=12)
plt.xlabel('Age', fontsize=12)
```
